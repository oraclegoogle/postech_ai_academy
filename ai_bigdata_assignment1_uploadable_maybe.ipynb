```python


#code = [20180001, 20180002, 20180007, 20180009, 20180011]
#name = ["Hong Gildong", "Lee Jieun", "Kim Cheolsu", "Lee Yeonghee", "Ha Donghun"]
#mid = [84, 92, 57, 81, 58]
#fin = [73, 89, 62, 84, 68]
#result = []
#rank = []
code = []
name = []
mid = []
fin = []
result = []
rank = []

#import sys
# 파일 읽기 (기본: students.txt)
filename = "students.txt" if len(sys.argv) < 2 else sys.argv[1]

with open(filename, "r") as f:
    lines = f.readlines()

for line in lines:
    parts = line.strip().split("\t")  # 탭으로 구분된 데이터
    student_code = parts[0]
    student_name = " ".join(parts[1:-2])
    mid_score = int(parts[-2])
    fin_score = int(parts[-1])
    avg_score = (mid_score + fin_score) / 2
    grade = calculate_grade(avg_score)

    code.append(student_code)
    name.append(student_name)
    mid.append(mid_score)
    fin.append(fin_score)
    result.append(avg_score)
    rank.append(grade)



# rank 산정 함수, avg_score를 입력하여서 각 학생별 rank를 판정함.
def calculate_grade(avg_score):
    if avg_score >= 90:
        return 'A'
    elif avg_score >= 80:
        return 'B'
    elif avg_score >= 70:
        return 'C'
    elif avg_score >= 60:
        return 'D'
    else:
        return 'F'

# 학점계산방법 함수.
def update_scores():
    result.clear()
    rank.clear()
    for i in range(len(code)):
        avg_score = (mid[i] + fin[i]) / 2
        result.append(avg_score)
        rank.append(calculate_grade(avg_score))

# 평균 및 학점 업데이트 선언.
update_scores()

# 리스트 컴프리핸션을 이용해서 내림차순으로 학생들을 정렬하기, x[4]가 평균점수이므로
def sort_students():
    global code, name, mid, fin, result, rank
    students = list(zip(code, name, mid, fin, result, rank))
    students.sort(key=lambda x: x[4], reverse=True)
    code, name, mid, fin, result, rank = zip(*students)
    code, name, mid, fin, result, rank = list(code), list(name), list(mid), list(fin), list(result), list(rank)

sort_students()

# show 함수
def show():
    print("Student          Name         Midterm   Final  Average  Grade")
    print("-------------------------------------------------------------")
    for i in range(len(code)):
        print(f"{code[i]:<12} {name[i]:<18} {mid[i]:<8} {fin[i]:<6} {result[i]:<8.1f} {rank[i]}")
    print()
    print()
    print("#")

# search 함수
def search():
    student_id = int(input("Student ID: "))
    if student_id in code:
        idx = code.index(student_id)
        print("Student          Name         Midterm   Final  Average  Grade")
        print("-------------------------------------------------------------")
        print(f"{code[idx]:<12} {name[idx]:<18} {mid[idx]:<8} {fin[idx]:<6} {result[idx]:<8.1f} {rank[idx]}") #위의 format과 위치를 일치시키기위해 미리 자리길이를 제한함.
    else:
        print("NO SUCH PERSON.")
    print()

# changescore 함수

def changescore():
    student_id = int(input("Student ID: "))
    if student_id not in code:
        print("NO SUCH PERSON.\n")
        return

    idx = code.index(student_id)
    category = input("Mid/Final? ").lower()
    if category not in ["mid", "final"]:
        print()
        return

    new_score = int(input("Input new score: "))

    # 점수 범위 검사 후 잘못된 경우 즉시 종료
    if not (0 <= new_score <= 100):
        print()
        return

    print("Student          Name         Midterm   Final  Average  Grade")
    print("-------------------------------------------------------------")
    print(f"{code[idx]:<12} {name[idx]:<18} {mid[idx]:<8} {fin[idx]:<6} {result[idx]:<8.1f} {rank[idx]}")


    if category == "mid":
        mid[idx] = new_score
    else:
        fin[idx] = new_score

    update_scores()
    sort_students()

    print("Score changed.")

    idx = code.index(student_id)
    print(f"{code[idx]:<12} {name[idx]:<18} {mid[idx]:<8} {fin[idx]:<6} {result[idx]:<8.1f} {rank[idx]}")



# add 함수
def add():
    student_id = int(input("Student ID: "))
    if student_id in code:
        print("ALREADY EXISTS.")
        print()
        return
    student_name = input("Name: ")
    mid_score = int(input("Midterm Score: "))
    fin_score = int(input("Final Score: "))
    code.append(student_id)
    name.append(student_name)
    mid.append(mid_score)
    fin.append(fin_score)
    update_scores()
    sort_students()
    print("Student added.")
    print()


# searchgrade 함수
def searchgrade():
    grade = input("Grade to search: ").upper()
    if grade not in ["A", "B", "C", "D", "F"]:
        print()
        return
    filtered_students = [(code[i], name[i], mid[i], fin[i], result[i], rank[i]) for i in range(len(rank)) if rank[i] == grade]
    if not filtered_students:
        print("NO RESULTS.")
        print()
        return
    else:
        print("Student          Name         Midterm   Final  Average  Grade")
        print("-------------------------------------------------------------")
    for student in filtered_students:

        print(f"{student[0]:<12} {student[1]:<18} {student[2]:<8} {student[3]:<6} {student[4]:<8.1f} {student[5]}")

    print("#")

# remove 함수
def remove():
    if not code:
        print("List is empty.")
        print()
        print("#")
        return
    student_id = int(input("Student ID: "))
    if student_id not in code:
        print("NO SUCH PERSON.")
        print()
        return
    idx = code.index(student_id)
    code.pop(idx)
    name.pop(idx)
    mid.pop(idx)
    fin.pop(idx)
    result.pop(idx)
    rank.pop(idx)
    print("Student removed.")
    print()

# quit 함수
def quit():
    save = input("Save data?[yes/no] ").lower()
    if save == "yes":
        filename = input("File name: ")
        with open(filename, "w") as f:
            for i in range(len(code)):
                f.write(f"{code[i]}\t{name[i]}\t{mid[i]}\t{fin[i]}\n")
        print("record system quit!")
    else:
      print("record system is still activated...")
    return

# main함수
show()
while True:
    command = input("# ").strip().lower()
    if command == "show":
        show()
    elif command == "search":
        search()
    elif command == "changescore":
        changescore()
    elif command == "add":
        add()
    elif command == "searchgrade":
        searchgrade()
    elif command == "remove":
        remove()
    elif command == "quit":
        quit()
        continue
    else:
        print("wrong approach! please check the available options!(show/search/changescore/add/searchgrade/remove/quit)")

```
